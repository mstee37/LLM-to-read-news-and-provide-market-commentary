```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def mineNews_OuputDataFrame(input, num_pages):

    df = pd.DataFrame(columns=[input])
    hashset = set()
    
    start = 0
    
    for page in range(num_pages):
    
        start = page * 10 # page 1: start=0, page 2: start=10, page 3: start=20
    
        # URL of the website to scrape
        url = f"https://www.google.com/search?q={input}&sca_esv=c7a01837b1eb8e0f&rlz=1C1UEAD_enSG1060SG1060&tbs=qdr:d,sbd:1&tbm=nws&sxsrf=ADLYWIJuzOuTZOex_My9k7KzN3z2MAPdSA:1725633163846&ei=ixLbZpmnM8jD4-EPxezCKQ&start={start}&sa=N&ved=2ahUKEwjZmPm9xK6IAxXI4TgGHUW2MAUQ8NMDegQIAxAS&biw=922&bih=908&dpr=1"
        
        # Send a GET request to the webpage
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        # Find all the news titles in the webpage (this depends on the webpage's structure)
        contents = soup.find_all("div")  # Adjust based on the correct HTML tag

        for x in contents:
            content = x.text.strip()
            
            list_content = content.split()
            is_duplicate = False  # Flag to check if content is duplicate

            # Check if the first 5 words of content already exist in the hashset
            for existing_content in hashset:
                list_existing = existing_content.split()
                
                # Debug print statements to see the comparison
                # print(f'list_existing = {list_existing[:5]}')
                # print(f'list_content = {list_content[:5]}')

                # Check if the first 5 words match exactly
                if list_content[:5] == list_existing[:5]:
                    is_duplicate = True
                    break  # Exit the loop if a match is found

            # If the content is a duplicate, skip it
            if is_duplicate:
                continue
            
            # If content is unique, add to hashset
            # print(content)
            new_row = pd.DataFrame({input: [content]})
            df = pd.concat([df, new_row], ignore_index=True)
            hashset.add(content)

    return df
```

```python
list = ["financial+markets+news", "fixed+income+news", "currencies+news", "equities+news", "commodities+news"]
# list = ["global+news+today"]
df = pd.DataFrame()
for x in list:
    print(x)
    df = pd.concat([df, mineNews_OuputDataFrame(x, 5)], axis=1, join="outer")

# print(df)
df.to_csv("google_news.csv", index=False)
df
```
